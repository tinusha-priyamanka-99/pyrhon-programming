```python
from tkinter import *
from tkinter.ttk import *

#create window
window=Tk()
window.geometry("320x160")
window.title("My Name")
window.resizable(0,0)
window.configure(background="white")

#create frames
fr1=Frame(window)
fr1.pack()
fr2=Frame(window)
fr2.pack()
fr3=Frame(window)
fr3.pack()
fr4=Frame(window)
fr4.pack()
fr5=Frame(window)
fr5.pack()
fr6=Frame(window)
fr6.pack()
fr7=Frame(window)
fr7.pack(side="bottom")

#variables
ex=""
color=""

#frame 1   
full_name = StringVar()   
lb1=Label(fr1,text="s14948",foreground="blue",background="blue",width=50)
lb1.pack(side="left")

#mouse event display hidden index number 
def enter_1(event):
    lb1.configure(background="cyan")    
def leave_1(event):
    lb1.configure(background="blue")

lb1.bind("<Enter>",enter_1)
lb1.bind("<Leave>",leave_1)

#frame2
en1=Entry(fr2,background="white",width=50,textvariable=full_name,justify="right")
en1.pack(side="left")
    
#display given input in entry    
def clicked(name,full_name):
    global ex
    ex=ex + name
    full_name.set(ex)
    
#mouse event reset given input in entry     
def reset(event):
    global ex
    global full_name
    en1.configure(foreground="black")
    ex=""
    full_name.set("")

en1.bind("<Button-3>",reset)

#reversed given input in entry     
def reverse():
    global ex
    ex=''.join(reversed(ex))
    full_name.set(ex)
    
#frame3    
bt1=Button(fr3,text="t",width=6,command=lambda:clicked("t",full_name))
bt1.pack(side="left")
bt1=Button(fr3,text="i",width=6,command=lambda:clicked("i",full_name))
bt1.pack(side="left")
bt1=Button(fr3,text="n",width=6,command=lambda:clicked("n",full_name))
bt1.pack(side="left")
bt1=Button(fr3,text="u",width=6,command=lambda:clicked("u",full_name))
bt1.pack(side="left")
bt1=Button(fr3,text="s",width=6,command=lambda:clicked("s",full_name))
bt1.pack(side="left")
bt1=Button(fr3,text="h",width=6,command=lambda:clicked("h",full_name))
bt1.pack(side="left")

#frame4
bt1=Button(fr4,text="p",width=6,command=lambda:clicked("p",full_name))
bt1.pack(side="left")
bt1=Button(fr4,text="r",width=6,command=lambda:clicked("r",full_name))
bt1.pack(side="left")
bt1=Button(fr4,text="y",width=6,command=lambda:clicked("y",full_name))
bt1.pack(side="left")
bt1=Button(fr4,text="m",width=6,command=lambda:clicked("m",full_name))
bt1.pack(side="left")
bt1=Button(fr4,text="k",width=6,command=lambda:clicked("k",full_name))
bt1.pack(side="right")
bt1=Button(fr4,text="a",width=6,command=lambda:clicked("a",full_name))
bt1.pack(side="right")

#frame5
bt1=Button(fr5,text="space",width=10,command=lambda:clicked(" ",full_name))
bt1.pack(side="left")
bt1=Button(fr5,text="reverse",width=10,command=lambda:reverse())
bt1.pack(side="right")
  
#frame6
lb2=Label(fr6,text="Tinusha Priyamanka",foreground="white",background="white",width=50)
lb2.pack(side="left")

#mouse event display hidden name    
def enter_2(event):
    lb2.configure(background="black")    
def leave_2(event):
    lb2.configure(background="white")

lb2.bind("<Enter>",enter_2)
lb2.bind("<Leave>",leave_2)

#frame7
lb3=Label(fr7,text="Select the Display Color:",background="white")
lb3.pack(side="left")
lb4=Label(fr7,background="red",width=5)
lb4.pack(side="left")
lb5=Label(fr7,background="blue",width=5)
lb5.pack(side="left")
lb6=Label(fr7,background="orange",width=5)
lb6.pack(side="left")
lb7=Label(fr7,background="green",width=5)
lb7.pack(side="left")

#mouse event changing color in entry    
def redfill(event):
    global color
    color="red"
    en1.configure(foreground=color)
def bluefill(event):
    global color
    color="blue"
    en1.configure(foreground=color)
def orangefill(event):
    global color
    color="orange"
    en1.configure(foreground=color)
def greenfill(event):
    global color
    color="green"
    en1.configure(foreground=color)

lb4.bind("<Button-1>",redfill)
lb5.bind("<Button-1>",bluefill)
lb6.bind("<Button-1>",orangefill)
lb7.bind("<Button-1>",greenfill)
      
window.mainloop()
```
